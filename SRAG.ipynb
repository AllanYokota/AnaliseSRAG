```python
path = '/content/drive/MyDrive/Colab Notebooks/INFLUD20-01-05-2023.csv'

df1 = pd.read_csv(
    path,                   # caminho do arquivo a ser lido 
    encoding='iso-8859-1',  # codificação dos caracteres 
    sep=';',                # separador de colunas
    usecols=['NU_IDADE_N', 'EVOLUCAO'] # IMPORTANTE: ler só a coluna TP_COR_RACA
)

path = '/content/drive/MyDrive/Colab Notebooks/INFLUD21-01-05-2023.csv'

df2 = pd.read_csv(
    path,                   # caminho do arquivo a ser lido 
    encoding='iso-8859-1',  # codificação dos caracteres 
    sep=';',                # separador de colunas
    usecols=['NU_IDADE_N', 'EVOLUCAO'] # IMPORTANTE: ler só a coluna TP_COR_RACA
)

path = '/content/drive/MyDrive/Colab Notebooks/INFLUD22-03-04-2023.csv'

df3 = pd.read_csv(
    path,                   # caminho do arquivo a ser lido 
    encoding='iso-8859-1',  # codificação dos caracteres 
    sep=';',                # separador de colunas
    usecols=['NU_IDADE_N', 'EVOLUCAO'] # IMPORTANTE: ler só a coluna TP_COR_RACA
)

path = '/content/drive/MyDrive/Colab Notebooks/INFLUD23-03-04-2023.csv'

df4 = pd.read_csv(
    path,                   # caminho do arquivo a ser lido 
    encoding='iso-8859-1',  # codificação dos caracteres 
    sep=';',                # separador de colunas
    usecols=['NU_IDADE_N', 'EVOLUCAO'] # IMPORTANTE: ler só a coluna TP_COR_RACA
)

df = pd.concat([df1, df2, df3, df4])
```

```python
totalcasosObitos = df[(df.NU_IDADE_N <=100) & (df.NU_IDADE_N >= 0)].groupby(by = ['NU_IDADE_N']).count()

df_temp = df[(df.NU_IDADE_N <=100) & (df.NU_IDADE_N >= 0) & (df.EVOLUCAO == 2) ].groupby(by = ['NU_IDADE_N'], as_index = False).count()
df_temp['EVOLUCAO'] = df_temp['EVOLUCAO'] / (totalcasosObitos['EVOLUCAO']/100)

sns.set(rc={'axes.facecolor':'#e4e4e4', 'figure.facecolor':'#e4e4e4'})

fig,ax = plt.subplots()

sns.lineplot(
    data=df_temp,
    x = 'NU_IDADE_N',
    y = 'EVOLUCAO',
    color = '#c3512c',
    dashes = False,
)

plt.grid('on', color = '#C6C6C6')

for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_edgecolor('#C6C6C6')

plt.title(
    'Taxa de mortalidade por idade\n',
    fontsize = 14,
    fontweight = 'bold',
    color = '#3C3C3C'
)

plt.xlabel(
    '\nIdade [Anos]',
    fontsize = 12,
    fontweight = 'bold',
    color = '#3C3C3C'
)

plt.yticks(
    color = '#3C3C3C'
)


plt.xticks(
    range(0,110,10),
    color = '#3C3C3C'
)

plt.ylabel(
    '\nTaxa de mortalidade [%]',
    fontsize = 12,
    fontweight = 'bold',
    color = '#3C3C3C'
)
```
